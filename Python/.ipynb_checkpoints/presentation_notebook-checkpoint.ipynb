On note $\pi_{\text{BTC}}$ le prix actuel de la Bitcoin et $\pi_{\text{BTC}}$ le nombre de bitcoin reçu à la découverte d'un bloc. La récompense vaut

$$b = n_{\text{BTC}}\cdot\pi_{\text{BTC}}$$



```python
Date = '2020-11-12'
# Importation of the time series
π_BTC_data = pd.read_csv("BTC_USD_2013-10-01_2020-11-13-CoinDesk.csv")
# Date reformating
π_BTC_data['Date2'] = pd.to_datetime(π_BTC_data['Date'])
π_BTC_df = π_BTC_data[['Date2','Closing Price (USD)']].rename(columns={"Date2": "Date", "Closing Price (USD)": "π_BTC"})

print('On November 12, 2020') 
# Block reward
n_BTC, π_BTC = 6.25, π_BTC_df.π_BTC[π_BTC_df.Date == Date].values[0]
b = n_BTC * π_BTC
print('The BTC price is $' + str(π_BTC) )
print('The reward for finding a block amount to $' + str(b) )
```

La consomation électrique estimée par le Cambridge Bitcoin Electricity Consumption Index [CBECI](https://cbeci.org/) en TWh par an est rapporté au kWh par heure et noté $W$.

```python
tW_data = pd.read_csv("elec_consumption_BTC_CBECI_update.csv")
tW_data['Date'] = pd.to_datetime(tW_data['Date and Time'])
tW_df = tW_data[['Date','GUESS']].rename(columns={"GUESS": "tW"})
# Operational cost of the network
W =  tW_df.tW[tW_df.Date == Date].values[0] * 10**9 / 365.25 / 24

print('On November 12, 2020')
print('The network consumes ' + str(tW_df.tW[tW_df.Date == Date].values[0]) + ' tWh per year')  
print('The network consumes ' + str(W) + ' kWh per hour')  
```

L'unité de temps est l'heure, le réseau découvre des blocs à raison de 6 blocs par heure. Le nombre de bloc jusqu'à l'instant $t$ est modélisé par un processus de Poisson $(N_t)_{t\geq0}$ d'intensité $\lambda = 6$.

```python
# Intensity of the Poisson process
λ = 6
print('The time unit is the hour, the Poisson process intensity is λ = ' + str(λ))
```

La richesse de Sam est modélisée par le processus suivant 

$$
R_t = u-c\cdot t+b\cdot \tilde{N}_t,
$$

où 

- $u$ désigne la réserve initiale
- $c$ correspond au coût opérationel de minage
- $b$ est la récompense lors de la découverte d'un bloc
- $\tilde{N}_t$ est un processus de Poisson d'intensité $p\lambda$ égale au nombre de blocs trouvé par Sam jusqu'à l'instant $t\geq0$

$p\in(0,1/2)$ correspond à la proportion de la puissance de calcul du réseau détenu par Sam. Le coût opérationnel pour Sam est supposé égal à 

$$
c = p\cdot W\cdot \pi_{\text{W}}, 
$$

avec $\pi_{W}$ le prix du kWh d'électricité. La condition de rentabilité en moyenne par unité de temps s'écrit 

$$
b\cdot\lambda -  W\cdot \pi_{\text{W}}>0
$$

et ne dépend pas de $p$. (La rentabilité ne dépend pas de la puissance de calcul du mineur)

```python
print('The net profit condition holds if the cost of electricity is lower than '
      + str(np.around(λ * b / W, decimals = 3))+
      ' when following the protocol')
```

Le temps de ruine est défini par 

$$
\tau_u = \inf\{t\geq0\text{ ; }R_t\leq0\}.
$$

La probabilité de ruine à horizon de temps $t\geq0$ est définie par 

$$
\psi(u,t) = \mathbb{P}(\tau_u \leq t).
$$

La probabilité de ruine ultime est définie par 

$$
\psi(u) = \mathbb{P}(\tau_u < \infty)
$$

Finalement, la richesse espérée sous contrainte de solvabilité est donnée par 

$$
V(u,t) = \mathbb{E}(R_t\mathbb{I}_{\tau_u>t}).
$$

La probabilités de ruine et la richesse espérée à horizon de temps exponentiel $T\sim \text{Exp}(t)$, notées

$$
\widehat{\psi}(u,t) = \mathbb{E}[\psi(u,T)],\text{ et } \widehat{V}(u,t) = \mathbb{E}[V(u,T)].
$$

Le but de la première illustration est de vérifier la proximité entre $\psi(u,t)$ et $\widehat{\psi}(u,t)$, et $V(u,t)$ et $\widehat{V}(u,t)$. Les paramètres sont 

$$
t = 6h,\pi_W = 0.06,\text{ and } p = 0.1.
$$

```python
# Time horizon, price of the electricity and  hashpower
t, π_W, p = 6, 0.06, 0.1
# operational cost per hour
c = p * π_W * W
u_vec = np.arange(1000, 510000, 1000) # Sequence of initial reserves
```

```python
ψ_u_t_vec = [ψ_t(u, t, b, c, p, λ) for u in u_vec]
ψ_u_T_vec = [ψ_T(u, t, b, c, p, λ) for u in u_vec]
ψ_u_vec = [ψ(u, b, c, p, λ) for u in u_vec]

plt.plot(u_vec, ψ_u_vec, '-', label = '$\psi(u)$', color = 'k')
plt.plot(u_vec, ψ_u_t_vec, '--', label = '$\psi(u,6h)$', color = 'k', linewidth=1)
plt.plot(u_vec, ψ_u_T_vec, linestyle = 'dotted', label = '$\widehat{\psi}(u,6h)$', color = 'k', linewidth=1)
plt.xlabel('u (réserve initiale en USD)', fontsize = fontsize_axes)
plt.title('Probabilité de ruine', fontsize = fontsize_axes)
plt.tick_params(labelsize=fontsize_ticker)
plt.legend(fontsize=fontsize_ticker)
```

Une analyse de sensitivité aux paramètres est menée en calculant la probabilité de ruine pour différente valeur de $u,t,\pi_W$ et $p$. 

```python
u_vec = np.arange(10000, 2050000, 50000)
t_vec = np.array([24, 7*24, 14*24])
π_W_vec = np.arange(0.04, 0.1002, 0.002)
p_vec = np.arange(0.01, 0.21, 0.01)
ψ_V_df = pd.DataFrame({'u':[],'t':[],'π_W':[], 'p':[], 'ψ_u_t':[], 'ψ_u':[],'V_u_T':[], 'EG':[]})
for t in t_vec:
    for π_W in π_W_vec:
        for p in p_vec:
            c = p * π_W * W
            res = pd.DataFrame({'u':u_vec,
                        't':np.repeat(t, len(u_vec)), 
                        'π_W':np.repeat(π_W, len(u_vec)),
                        'p':np.repeat(p, len(u_vec)),
                        'ψ_u_t':[ψ_t(u, t, b, c, p, λ) for u in u_vec], 
                        'ψ_u_T':[ψ_T(u, t, b, c, p, λ) for u in u_vec],
                        'ψ_u':[ψ(u, b, c, p, λ) for u in u_vec], 
                        'V_u_T':[ V_T(u, t, b, c, p, λ) for u in u_vec],
                       'EG': np.repeat(- c * t + p * λ * b * t, len(u_vec))})
            ψ_V_df = pd.concat([ψ_V_df, res])
ψ_V_df
```

## The impact of joining a mining pool on the risk reduction

Soient $100$ mineurs possédant chacun $u = \$10,000$ et une proportion $p = 0.01$ de la puissance de calcul. La formation d'un pool de deux mineurs conduit a l'addition des réserves initiales $ u = \$10,000$ et des puissances de calcul $p  = 0.02$. Nous nous intéressons à l'effet sur le profit d'un mineur au sein d'un groupe de taille croissante. Le prix de l'électricité est fixé à $\pi_W = 0.06$.

```python
n_miner_vec = np.arange(1,101,1)
t_vec = np.array([24, 7*24, 14*24])
u = 10000
p_miner= 0.01
π_W = 0.06
ψ_V_pool = pd.DataFrame({'n_miner':[],'t':[], 'ψ_u_t':[], 'V_u_T':[]})
for t in t_vec:
    res = pd.DataFrame({'n_miner':n_miner_vec,
                        't':np.repeat(t, len(n_miner_vec)),
                        'ψ_u_t':np.array([ψ_t(n_miner *u, t, b, n_miner * p_miner * π_W * W,  n_miner * p_miner, λ) 
                                          for n_miner in n_miner_vec]),
                        'ψ_u_T':np.array([ψ_t(n_miner *u, t, b, n_miner * p_miner * π_W * W,  n_miner * p_miner, λ) 
                                          for n_miner in n_miner_vec]),
                        'V_u_T':np.array([V_T(n_miner *u, t, b, n_miner * p_miner * π_W * W,  n_miner * p_miner, λ) / n_miner 
                                          for n_miner in n_miner_vec])
                       })
    ψ_V_pool = pd.concat([ψ_V_pool, res])
```

### Profit espéré et probabilité de ruine en fonction de la taille du pool de minage pour différents horizons de temps

```python
linestyles = ['dotted', 'dashed', 'dashdot']
t_names = np.array(['1d', '1w', '2w'])

for k in range(len(t_vec)):
    plt.plot(n_miner_vec, (ψ_V_pool.V_u_T[np.around(ψ_V_pool.t.values, decimals = 0) == t_vec[k]].values - u),
             color = 'k', alpha = 0.5,linestyle = linestyles[k], label = '$t='+t_names[k]+'$')

plt.xlabel('number of miners', fontsize = fontsize_axes)
plt.title('Expected profit of each miner', fontsize = fontsize_axes)
plt.tick_params(labelsize=fontsize_ticker)
plt.legend(fontsize = fontsize_axes)
```
